## Matrix Multiplication <a name='matrix-multiplication'> </a>
* [Dot Product](#dot-product)
* [Cross Product](#cross-product)

### Dot-Product <a name='dot-product'> </a>

````python
# '@': does matrix dot product 
# similar to '.dot()' function provided by numpy
print(f'Array1 @ Array2: {basic_numpy_array1@basic_numpy_array2}')
print('-'*20)
print(f'2dArray1 @ 2dArray2:\n{two_dimensional_array1@two_dimensional_array2}')
print('-'*20)
````

### Cross Product <a name='cross-product'></a>



````python
# '.cross()': does matrix cross product
# only for 2 or 3 dimensional arrays
#print(f'Array1 * Array2: {np.cross(basic_numpy_array1,basic_numpy_array2)}')
#print('-'*20)
# the above line generates an error
print(f'np.cross(2dArray1,2dArray2):\n{np.cross(two_dimensional_array1,two_dimensional_array2)}')
print('-'*20)
````

### Element-wise Multiplication <a name='element-multiplication'></a>



````python
# '*': does matrix element-wise multiplication
# similar to '.multiply()' function provided by numpy
print(f'Array1 * Array2: {basic_numpy_array1*basic_numpy_array2}')
print('-'*20)
print(f'2dArray1 * 2dArray2:\n{two_dimensional_array1*two_dimensional_array2}')
print('-'*20)
````

## Matrix Addition <a name='matrix-addition'> </a>

````python
# '+': simply adds the 2 given arrays
# it also broadcasts (adds an integer value to an array)
print(f'Array1 + Array2: {basic_numpy_array1+basic_numpy_array2}')
print('-'*20)
print(f'Broadcasting of Array1 with int(5): {basic_numpy_array1+5}')
print('-'*20)
#print(f'Broadcasting of Array1 with another array([4,5,6]): {basic_numpy_array1+np.array([4,5,6])}')
#print('-'*20)
print(f'2dArray1 + 2dArray2:\n{two_dimensional_array1+two_dimensional_array2}')
print('-'*20)
print(f'Broadcasting of 2dArray1 with int(5):\n{two_dimensional_array1+5}')
print('-'*20)
print(f'Broadcasting of 2dArray1 with another array([4,5]):\n{two_dimensional_array1+np.array([4,5])}')
print('-'*20)
````

## Matrix Subtraction <a name='matrix-subtraction'> </a>

````python
# '-': simply subtracts the 2 given arrays
# similar as '.subtract()' function provided by numpy
print(f'Array1 - Array2: {basic_numpy_array1-basic_numpy_array2}')
print('-'*20)
print(f'2dArray1 - 2dArray2:\n{two_dimensional_array1-two_dimensional_array2}')
print('-'*20)
````

## Element Sumation <a name='element-sum'> </a>


````python
# '.sum()': simply add all the elements of the given array
print(f'Sum of Array1: {np.sum(basic_numpy_array1)}')
print('-'*20)
print(f'Sum of 2dArray1: {np.sum(two_dimensional_array1)}')
print('-'*20)
````

## Matrix Division <a name='matrix-div'> </a>

````python
# '.divide()': simply divides the 2 given arrays
# Syntax: np.divide(divident_array,divisor_array)
print(f'Dividing Array1 and Array2: {np.divide(basic_numpy_array1,basic_numpy_array2)}')
print('-'*20)
print(f'Dividing Array1 from int(5): {np.divide(basic_numpy_array1,5)}')
print('-'*20)
print(f'Floor-Dividing Array1 from int(5): {np.floor_divide(basic_numpy_array1,5)}')
print('-'*20)
print(f'Dividing 2dArray1 and 2dArray2:\n{np.divide(two_dimensional_array1,two_dimensional_array2)}')
print('-'*20)
print(f'Dividing 2dArray1 from int(5):\n{np.divide(two_dimensional_array1,5)}')
print('-'*20)
print(f'Floor-Dividing 2dArray1 and 2dArray2:\n{np.floor_divide(two_dimensional_array1,two_dimensional_array2)}')
print('-'*20)
````

### Numpy Random Function <a name='numpy-random'></a>

````python
normal_distribution = np.random.standard_normal(size=(3,4))
print(f'Normal Distribution:\n{normal_distribution}')
print('-'*20)
uniform_distribution = np.random.uniform(1,12,(3,4))
print(f'Uniform Distribution:\n{uniform_distribution}')
print('-'*20)
random_int_array = np.random.randint(1,51,(3,4))
random_float_array = np.random.rand()
print(f'Random int array:\n{random_int_array}')
print('-'*20)
print(f'Random float array:\n{random_float_array}')
print('-'*20)
````

### Numpy Zero array and One array <a name='numpy-0-1'></a>

````python
zero_array = np.zeros((2,3))
one_array = np.ones((3,2))
print(f'0 array:\n{zero_array}')
print('-'*20)
print(f'1 array:\n{one_array}')
print('-'*20)
````

### Numpy Logical Operations <a name='numpy-logical-ops'></a>

````python
# '.logical_<operation>()': performs logical operations on numpy arrays
filter_array1 = np.logical_and(random_int_array>20,random_int_array<50)
print(f'Filter Array 1:(Random array1 has values greater than 20 and less than 50)\n{filter_array1}')
filter_values = random_int_array[filter_array1]
print(f'Filter Values:\n{filter_values}')
````

## Getting Statistical Data <a name='numpy-stats'></a>

````python
print(f'Example Array: {basic_numpy_array1}')
print('-'*20)
mean = np.mean(basic_numpy_array1)
# '.mean()': calculates mean of array
variance = np.var(basic_numpy_array1)
# '.var()': calculates variance of array
std = np.std(basic_numpy_array1)
# '.std()': calculates standard deviation of array
median = np.median(basic_numpy_array1)
# '.median()': calculates median of array
print(f'Mean: {mean}')
print(f'Variance: {variance}')
print(f'Standard Deviation: {std}')
print(f'Median: {median}')
print('-'*20)
````

# Pandas Module and Examples <a name='pandas-code'> </a>

## Introduction <a name='pd-intro'></a>
**Pandas (Panel Data)** is an open-source library that is designed to work with DataFrames(multi-dimensional arrays) and Series (one-dimensional arrays) of data.

````python
# Author: Praman Kasliwal
# Version: 1.0 (Alpha)
# Title: Pandas Coding
# basically imports and variable declaration

import pandas as pd
import numpy as np
import random

# set pandas formatting for displaying all rows and columns without hiding any...
pd.set_option('display.max_columns',None)
pd.set_option('display.max_rows',None)

# set numpy formatting for float to 2 decimal places by default
np.set_printoptions(formatter={'float': lambda x: "{0:0.2f}".format(x)})

# ----------------------------------------------- #
# creating student records (ID, Age, Score)
s_id = pd.Series(random.sample([x for x in range(2000,5000)],50))
s_age = pd.Series(np.random.randint(15,30,size=50))
s_score = pd.Series(np.random.randint(20,100,size=50))
# ----------------------------------------------- #
````

## Series <a name='pd-srs'> </a>

````python
trialSeries = pd.Series([10,20,30,40],index=['a1','a2','a3','a4'])
print(f'Index a2 of Series: {trialSeries.a2}')
print('-'*20)
print(f'Filtering int value greater than 10:\n{trialSeries[trialSeries>10]}')
print('-'*20)
print(f'Listing only values: {trialSeries.values}')
print('-'*20)
print(f'Listing only indices: {trialSeries.index}')
print('-'*20)
trialSeries.index = ['age1','age2','age3','age4']
print(f'New Indices: {trialSeries.index}')
print('-'*20)
````

## DataFrames <a name='pd-df'> </a>

````python
# ----------------------------------------------- #
# Imp Note:
# Here, the index values start with 1, but by default it starts from 0
# ----------------------------------------------- #
student_dataSet = pd.DataFrame({'ID':s_id.values,'Age':s_age.values,'Score':s_score.values},index=[x for x in range(1,51)])
#print(student_dataSet)

# ----------------------------------------------- #
# list all the values of the dataset where score is greater than 80
print(f'List of students whose score is greater than 80:\n{student_dataSet[student_dataSet["Score"]>80]}')
# ----------------------------------------------- #

while True:
  try:
    roll = int(input('Roll Number (Enter 0 to exit): '))
    if roll > 0 and roll < 51:
      print(f'Data of student (Roll Number {roll}):\n{student_dataSet.loc[roll]}')
    elif roll == 0:
      break
    else:
      print('Error: Roll Numbers range from 1 to 50 only!')
  except ValueError as e:
    print('Please enter numbers only')
print('Thank You!')
````

# Matplotlib <a name='mtpltlib'> </a>

## Setup <a name='mtplt-setup'> </a>

````python
# Author: Praman Kasliwal
# Version: 1.0 (Alpha)
# Title: Matplotlib Coding
# basically imports and variable declaration

import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt

# set pandas formatting for displaying all rows and columns without hiding any...
pd.set_option('display.max_columns',None)
pd.set_option('display.max_rows',None)

# set numpy formatting for float to 2 decimal places by default
np.set_printoptions(formatter={'float': lambda x: "{0:0.2f}".format(x)})

# ----------------------------------------------- #
# creating student records (ID, Age, Score)
s_id = pd.Series(random.sample([x for x in range(2000,5000)],50))
s_age = pd.Series(np.random.randint(15,30,size=50))
s_score = pd.Series(np.random.randint(20,100,size=50))
student_dataSet = pd.DataFrame({'ID':s_id,'Age':s_age,'Score':s_score},index=[x for x in range(51)])
# ----------------------------------------------- #

# ----------------------------------------------- #
# creating sample electricity consumption data for 2 customers
month = ['Jan','Feb','Mar','Apr','May','Jun',
         'Jul','Aug','Sep','Oct','Nov','Dec']
cust1 = [12,13,9,8,7,8,8,7,6,5,8,10]
cust2 = [14,16,11,7,6,6,7,6,5,8,9,12]
# ----------------------------------------------- #
print('Setup Complete')
````

## Line Graph <a name='mtplt-line'> </a>
Syntax: `plt.plot(X-Value,Y-Value,<other optional attributes>)`

````python
# Line Graph
id = [x for x in range(1,51)]
scores = s_score.values
#print(id,scores)
plt.plot(id,scores)
plt.xlabel('Student Roll Number')
plt.ylabel('Score Achieved')
plt.title('Student Records',{'fontsize':18,'horizontalalignment':'left'})
plt.show()
````

## Scatter Plot<a name='mtplt-scatter'> </a>
Syntax: `plt.scatter(X-Value,Y-Value,<other optional attributes>)`

````python
month = ['Jan','Feb','Mar','Apr','May','Jun',
         'Jul','Aug','Sep','Oct','Nov','Dec']
cust1 = [12,13,9,8,7,8,8,7,6,5,8,10]
cust2 = [14,16,11,7,6,6,7,6,5,8,9,12]

plt.scatter(month,cust1,color='red',label='Customer 1',marker='*')
plt.scatter(month,cust2,color='blue',label='Customer 2',marker='^')
plt.xlabel('Month')
plt.ylabel('Electricity Consumed (in kW)')
plt.title('Year 2001: Electricity Consumption')
plt.grid()
plt.legend()
plt.show()
````

## Sub-plot <a name='mtplt-sub-plot'> </a>

Syntax: `plt.subplot(row,column,graph_number)`
```
Where,
  row: Number of rows
  column: Number of columns
  graph_number: Self-explanatory
```


````python
plt.subplot(1,2,1)
plt.plot(month,cust1,color='red',label='Customer 1',marker='*')
plt.xlabel('Month')
plt.ylabel('Electricity Consumed (in kW)')
plt.title('Year 2001: Customer 1')
plt.show()
plt.subplot(1,2,2)
plt.plot(month,cust2,color='blue',label='Customer 1',marker='^')
plt.xlabel('Month')
plt.ylabel('Electricity Consumed (in kW)')
plt.title('Year 2001: Customer 2')
plt.show()
````

## Histogram <a name='mtplt-hist'> </a>
Shows Number of occurences of each of data. <br>
Syntax: `plt.hist(Y_values,bins=integer,<other optional attributes>)`

````python
plt.hist(cust1,bins=20,color='green')
plt.xlabel('Month')
plt.ylabel('Electricity Consumed (in kW)')
plt.title('Year 2001: Customer 1 (Histogram)')
plt.show()
````

## Bar Chart <a name='mtplt-bar'> </a>
Syntax: `plt.bar(X-Values,Y-Values,<other optional attributes>)`

````python
bar_width = 0.4
month_b = np.arange(12)
plt.bar(month_b,cust1,width=bar_width,color='b',label='Customer 1')
plt.bar(month_b+bar_width,cust2,width=bar_width,color='r',label='Customer 2')
plt.xlabel('Month')
plt.ylabel('Electricity Consumed (in kW)')
plt.title('Year 2001: Electricity Consumption')
plt.xticks(month_b+(bar_width)/2.5, tuple(month))
plt.legend()
plt.show()
````

## Pie Chart <a name='mtplt-pie'> </a>
Syntax: `plt.pie(list,<other optional attributes>)`

````python
more_than_60 = student_dataSet[student_dataSet['Score'] >= 60]
less_than_60 = student_dataSet[student_dataSet['Score'] < 60]
#print(less_than_60,'\n',more_than_60)
#print(len(less_than_60),len(more_than_60))
pcent_less = (len(less_than_60)/50)*100
pcent_more = (len(more_than_60)/50)*100
#print(pcent_less+pcent_more)
plt.pie([pcent_less,pcent_more],labels=['Less than 60%','60% or more'],startangle=90,shadow=True,)
plt.title('Student Percentage (who scored more than 60% vs who did not)')
plt.legend(loc='upper left')
plt.show()
````

## Box Plot <a name='mtplt-box'> </a>
Syntax: ` plt.boxplot() `

````python
#plt.boxplot(cust1,notch=False,vert=True)
plt.boxplot([cust1,cust2],patch_artist=True,
            boxprops=dict(facecolor='r',color='red'),
            whiskerprops=dict(color='g'),
            capprops=dict(color='m'),
            medianprops=dict(color='y'),
            notch=False,vert=True)
````
